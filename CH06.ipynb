# Indexing

```python
import pandas as pd
```

```python
oo = pd.read_csv('olympics.csv', skiprows=4)
oo.head()
```

### Index

```python
oo.index[500]
```

#### Immutable Example

```python
oo.index[500] = 5
```

### set_index()

```python
oo.head()
```

```python
oo.set_index('Athlete', inplace=True)
oo.head()
```

```python
oo.reset_index(inplace=True)
oo.head()
```

```python
ath = oo.set_index('Athlete')
ath.head()
```

### reset_index()

```python
ath.head()
```

```python
ath.reset_index(inplace=True)
ath.head()
```
