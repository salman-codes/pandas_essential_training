```python
import pandas as pd
```

```python
oo = pd.read_csv('olympics.csv', skiprows=4)
oo.head()
help(pd.read_excel('olympics.xlsx'))
```

### In which events did Jesse Owens win a medal?

```python
jo = oo[oo.Athlete =='OWENS, Jesse']
```

```python
jo.Event.value_counts()
```

### Which country has won the most men's gold medals in singles badminton over the years?

### Sort the result by Athlete name

```python
gbm = oo[(oo.Gender == 'Men') & (oo.Medal == 'Gold') & (oo.Sport == 'Badminton')]
gbm
```

```python
gbm.sort_values(by='Athlete',ascending=True)
```

### Which three countries have won the most medals in recent years (from 1984 to 2008)?

```python
rec = oo[oo.Edition >= 1984]
rec.NOC.value_counts().head(3)
```

### Display the male gold medal winners for the 100m track and field sprint event over the years.

#### List the result starting with the most recent.

#### Show the olympic, city, edition, athlete, and the country they represent.

```python
gmh = oo[(oo.Gender == 'Men') & (oo.Event == '100m') & (oo.Medal == 'Gold') ]
gmh.sort_values(by='Edition', ascending=False)[['City', 'Edition', 'Athlete', 'NOC']]
```
